```python

s['starttime'] = pd.to_datetime(s['starttime'])
s['stoptime'] = pd.to_datetime(s['stoptime'])
s['trip duration'] = (s['stoptime'] - s['starttime'])
print(s['trip duration'].mean().seconds)
```

```python
def fn(arg):
    if arg==5 or arg==6:
        return 1
    else:
        return 0
s['week']=s['starttime'].dt.dayofweek
s['week']=s['week'].apply(fn)
print(s[s['week']==1].shape[0])
s.head()
```

```python
def func(arg):
    if arg >=0 and arg<=6:
        return 'night' 
    elif arg >6 and arg<=12:
        return 'morning'
    elif arg >12 and arg<=18:
        return 'day'
    else:
        return 'evening'
s['time_of_day']=s['starttime'].dt.hour
s['time_of_day']=s['time_of_day'].apply(func)
dd=s[s['time_of_day']=='day'].shape[0]
nn=s[s['time_of_day']=='night'].shape[0]
print(round(dd/nn))
   
```
